# Fatorial
$$ n! = n(n-1)\dots2. 1 $$

```julia
function fatorial(n)
    resultado = 1
    for i = 1:n
        resultado = resultado * i
    end
    return resultado
end
```

```julia
fatorial(4)
```

```julia
fatorial(5)
```

```julia
fatorial(0)
```

```julia
function sinal(x)
    if x > 0
        return 1
    elseif x < 0
        return -1
    else
        return 0
    end
end
```

```julia
sinal(3.2)
```

```julia
sinal(-1.2)
```

```julia
sinal(0.0)
```

```julia
function ordem(N) # ordem(12) = 1, ordem(500) = 2
    n = 0
    while N > 10
        n = n + 1
        N = N/10
    end
    return n
end
```

```julia
ordem(12)
```

```julia
ordem(500)
```

```julia
ordem(-1) # NÃ£o funciona
```

```julia
x = linspace(0, 2*pi, 100)
anim = Animation()
for i = 1:3:100
    plot(x[1:i], sin(x[1:i]))
    xlims!(x[1], x[end])
    ylims!(-1, 1)
    frame(anim)
end
gif(anim, "exemplo2.gif", fps=12)
```

```julia
x = linspace(0, 2*pi, 100)
anim = Animation()
for i = 1:3:100
    plot(x, sin(x), leg=false)
    a = x[i]
    scatter!([a], [sin(a)], c=:red)
    plot!(x, sin(a) + cos(a)*(x-a), c=:red, l=:dash)
    xlims!(x[1], x[end])
    ylims!(-2, 2)
    frame(anim)
end
gif(anim, "exemplo3.gif", fps=12)
```
